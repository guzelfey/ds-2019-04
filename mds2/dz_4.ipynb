Дана параметрическая модель человека. 

*   Посчитать угол, который образуется от вектора позвоночника до вектора горизонтали.
*   Рассчитать рост человека.



![alt text](https://github.com/yustiks/3D_model_test_task/blob/master/snapshot01.png?raw=true)

```python
data = pd.read_csv('joints.txt', sep=" ", header=None)
data.columns = ["smt","x", "y", "z"]
data.drop(["smt"],axis=1, inplace=True)
data
```

```python
data.drop_duplicates(inplace=True)
data.sort_values(by=['x','y','z'],inplace=True)
data=data.reset_index()
data.drop(['index'],axis=1,inplace=True)
data
```

```python
data.describe()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(data['x'], data['y'], data['z'],cmap=plt.cm.viridis, linewidth=0.2)
ax.view_init(30, 210)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(data['x'], data['y'], data['z'],cmap=plt.cm.viridis, linewidth=0.2)
ax.view_init(90, 210)
plt.show()
```

```python
for angle in range(0,360,10):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.plot_trisurf(data['x'], data['y'], data['z'], cmap=plt.cm.viridis, linewidth=0.2)
    # Set the angle of the camera
    ax.view_init(30,angle)
    # Save it
    filename='\joints\joints'+str(angle)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()
```

Предположительно, человек лежит. Голова-ноги располагаются по оси y, руки - по оси x. <br>(Есть подозрение, что знак по z у двух точек торса может быть другой)

```python
#head
head_=np.array(data.iloc[6])
head_
```

```python
#hips
hips_=np.array(data.iloc[[5,11]].mean())
hips_
```

```python
#knees
knees_=np.array(data.iloc[[3,9]].mean())
knees_
```

```python
#feet
feet_=np.array(data.iloc[[4,10]].mean())
feet_
```

```python
vector_spine=head_-hips_
horiz=np.array([1,1,0])
```

Считаем угол между двумя векторами.

```python
math.degrees(math.acos(vector_spine.dot(horiz)/(np.linalg.norm(vector_spine)*np.linalg.norm(horiz))))
```

Посчитаем рост человека.

```python
np.linalg.norm(head_-hips_)+np.linalg.norm(hips_-knees_)+np.linalg.norm(knees_-feet_)
```
