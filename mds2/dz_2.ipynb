Загрузим датасет о ценах на дома в Бостоне. 

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.data.shape)
```

```python
boston.feature_names
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов и датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
x1.dot(x2)
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

ax.scatter(x1[2],x1[5],x1[12], c='b', marker='o')
ax.scatter(x2[2],x2[5],x2[12], c='r', marker='o')

ax.set_xlabel('INDUS')
ax.set_ylabel('NOX')
ax.set_zlabel('LSTAT')
plt.show()
```

Найти косинус угла между векторами x1 и x2

```python
x1.dot(x2)/(np.linalg.norm(x1)*np.linalg.norm(x2))
```

Найти расстояние между точками x1 и x2.

```python
np.linalg.norm(x1-x2)
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
B = np.array([[300,200,0]]*6)
C = A+B
plt.plot(A[:,0],A[:,1])
plt.plot(C[:,0],C[:,1])
```

Уменьшить объект в два раза.

```python
D = A/2
plt.plot(A[:,0],A[:,1])
plt.plot(D[:,0],D[:,1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов.

```python
E=np.array([[np.cos(2*np.pi/9),-np.sin(2*np.pi/9),0],[np.sin(2*np.pi/9),np.cos(2*np.pi/9),0],[0,0,1]])
F=A.dot(E)
plt.plot(A[:,0],A[:,1])
plt.plot(F[:,0],F[:,1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
G=np.array([[0,1,0],[1,0,0],[0,0,1]])
H=A.dot(G)
plt.plot(A[:,0],A[:,1])
plt.plot(H[:,0],H[:,1])
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
